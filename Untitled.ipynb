```python
import pandas as pd
import numpy as np
import scipy
import json
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df_thi19 = pd.read_csv("diemthi2019.csv")
df_thi20 = pd.read_csv("diemthi2020.csv")
```

```python
T = []
with open('university_diemchuan', encoding='utf-8') as f:
    for line in f.readlines():
        T.append(json.loads(line))
```

```python
D = []
for r in T:
    for v in r["diemchuan_datas"]:
        v["university_code"] = r["university_meta"]["university_code"]
        v["university_name"] = r["university_meta"]["university_name"]
        D.append(v)
```

```python
df_diemchuan = pd.DataFrame(D)
```

```python
fig, axs = plt.subplots(2,5)
fig.set_size_inches(18.5, 10.5)
for i,ax in enumerate(np.array(axs).reshape(-1)):
    col_name = df_thi19.columns[i + 1]
    if col_name == "Ma_mon_ngoai_ngu":
        continue
    elif col_name == "Ngoai_ngu":
        ax.hist(df_thi19[df_thi19["Ma_mon_ngoai_ngu"] == "N1"][col_name].dropna(), bins = "auto")
        ax.set_xlabel(col_name)        
    else:
        ax.hist(df_thi19[col_name].dropna(), bins = "auto")
        ax.set_xlabel(col_name)
fig.suptitle('Diem 2019', fontsize=20)
```
